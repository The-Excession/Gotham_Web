# Plotting 
This notebook goes over the beging of plotting in pythong and generationg data 

```python
# Let's begin by importing the packages we need 

import numpy as np 

import matplotlib.pyplot as plt #plotting package 

import seaborn as sn #plotting package 
```

```python
# To start we need some data to use in our graphs so let's begin with creating some numpy arrays 

x = np.linspace(0,10,100) #an array of values from 0 to 10 with 100 steps in between
print(x)
```

```python
# To get our y values lets create a function that computes y values given an x value

def y_calculation(x): 
    """
    This function will calculate y values given an x value or arrat of x values
    
    Input
    x : value or array  
    
    Output: 
    y: value or array
    
    """
    
    y= 4 + 1 * np.sin(2 * x)   # some function that rlates x and y       
                       
    return y 
# The return command allows us to use this value outside this function. The variable now becomes global. 
```

```python
# For plots to work you neeed the same number of y values as x's 

y = y_calculation(x) #calling our function 

print(y)
```

```python
plt.figure(facecolor='w', figsize=(10,8))

plt.plot(x,y) #the plot function will create a line between your points instead of a scatter 

plt.xlabel("X") #label on the x axis 

plt.ylabel("Y") #label on the y axis

plt.title("My Plot") # title label

plt.show()
```

### Customization
You can also custimize many features about this plotusing various keywords
* Many of these keywords can be found on the matplotlib page 

```python
#Custimized plot

plt.figure(facecolor='w', figsize=(10,8))

plt.plot(x,y,color='purple',linestyle= '--')

plt.xlabel("X",font= 'monospace', fontsize= 14,color='blue') #label on the x axis 

plt.ylabel("Y",font= 'monospace', fontsize= 14,color='red') #label on the y axis

plt.title("My Plot",font= 'monospace', fontsize= 18) # title label

plt.show()
```

### Differnt types of plots

There are various says of shoeing your data like scatter plots and historgrams here are some examples of these. 

```python
#Scatter PLot
plt.figure()
plt.scatter(x,y)
plt.xlabel("X2")
plt.ylabel("Y2")
plt.title("Scatter Plot")


#Historgram
plt.figure()
plt.hist(y) 
plt.title("Histogram of Y values")
plt.xlabel("Y - value")
plt.ylabel("Count")

#Notice that I didn't use the plt.show() function here
#this is becasue Jupyter notebook will automatically show these figures but this isn't the case when writting python scripts
```

### Plotting multiple things at once.

Sometimes you want to show multiple trends at once. To do this you can layer the plots


```python
#Frist we need to get our second array 

x2= np.linspace(-10,8,100)

y2= y_calculation(x2*.75)# the *.5 is just so these poionts will appear out of sync with our origional line
```

```python
# to add an aditional line all you have to do is call a plotting function within the same figure

plt.figure(figsize=(8,6))

plt.scatter(x,y,label="First Data Set",alpha=0.5)

plt.scatter(x2,y2,label="Second Data Set",alpha=0.5)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Two Dataset Plot")

plt.legend()
```

### Plotting Side by Side

To plot this side by side we need to introduce the axis and figure commands. When you are dealing with multiple plots you must indicate which axis you want the plot the be in withing the larger figure. 

```python
fig,(ax1,ax2,ax3) =plt.subplots(1,3,figsize=(40,10)) # the ax1,2,3 are for the axis of each subplot 
#the subplots function returns an axis for each subplot and larger figure

ax1.plot(x,y,color="steelblue") 

ax2.scatter(x2,y2,color= "purple")#

ax3.hist(y,color="steelblue", alpha =0.5,label="Y")
ax3.hist(y2,color="purple", alpha=0.5,label="Y2")
ax3.legend(fontsize=25)
 
#Becasue we are using ax instead of plt command, it uses slightly different commands for labeling 
ax1.set_xlabel("X",fontsize=30)
ax1.set_ylabel("Y",fontsize=30)
ax1.set_title("Subplot 1",fontsize=30)

ax2.set_xlabel("X2",fontsize=30)
ax2.set_ylabel("Y2",fontsize=30)
ax2.set_title("Subplot 2",fontsize=30)

ax3.set_xlabel("Y - Value",fontsize=30)
ax3.set_ylabel("Count",fontsize=30)
ax3.set_title("Subplot 3",fontsize=30)

#Because the title is part of the larger figure, we use fig to set the title
fig.suptitle("Multi Plot Figure",fontsize= 40)
```
